##Geração da matriz de palavras

```python
#Abrindo o arquivo no colab
with open('/content/Base Text - Matrix Representation I - v2.txt') as f:
    linhas = f.readlines()
    matriz = []
    for linha in linhas:         #Verificando cada linha da base
        if(linha != "\n"):
            palavras = linha.split()
        else:
            palavras = "\n"
        matriz.append(palavras)
```

```python
#visualizar as palavras da matriz de palavras
print(matriz)
```

##Texto original

```python
texto = ""
for i in matriz:
    linha = ""
    if(i == '\n'):
        linha = "\n"
    else:
        primeiro = True
        for j in i:
            if primeiro:
                linha = j
                primeiro = False
            else:
                linha = linha + " " + j
    if(linha == "\n"):
        texto = texto + linha
    else:
        texto = texto + linha + "\n"
with open('obter_texto_original.txt', 'w') as f:
    f.write(texto)
```

```python
#print do texto, para conferir com o texto original
print(texto)
```

#Salvar a matriz de palavras

```python
with open("Matriz_palavras", "wb") as fp: # Salvar o arquivo no formato pickle
  pickle.dump(matriz,fp)
```
